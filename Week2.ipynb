## 2. Tampilkan image dalam channel Red-Blue dan Green-Blue saja!

```python
img3 = cv.cvtColor(img,cv.COLOR_BGR2RGB)
plt.imshow(img3)

RB = img3[:, :, 0]  # Red-Blue
GB = img3[:, :, 1]  # Green-Blue

# Display the Red-Blue
plt.figure(figsize=(8, 4))
plt.subplot(1, 2, 1)
plt.imshow(RB, cmap='RdBu')
plt.title('Red-Blue')

# Display the Green-Blue
plt.subplot(1, 2, 2)
plt.imshow(GB, cmap='GnBu')
plt.title('Green-Blue')
```

## 3. Tampilkan image baris ke 20-115, kolom 25-120!
